## Making all columns lowercase and replacing dash with underscore

```python
df.columns = df.columns.str.strip().str.lower().str.replace('-', '_')

df.head()
```

## Switch schedule day and appointment to datetime and gender and neighbourhood to category

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
df['gender'] = df['gender'].astype('category')
df['neighbourhood'] = df['neighbourhood'].astype('category')
df.head()
```

```python
df.info()
```

## Days until the patient next appointment and creating week days and number of the month to calculate when is the time next patient appointment at and when did he schedule it

```python
df['days_until_appointment'] = (
    df['appointmentday'].dt.normalize() - df['scheduledday'].dt.normalize()
).dt.days


df['appointment_weekend'] = df['appointmentday'].dt.dayofweek >= 5


df['scheduled_month'] = df['scheduledday'].dt.month
df['scheduled_day_of_week'] = df['scheduledday'].dt.dayofweek
df['appointment_month'] = df['appointmentday'].dt.month
df['appointment_day_of_week'] = df['appointmentday'].dt.dayofweek

df.head()
```

```python
df.days_until_appointment
```

## To filter age from negative numbers and the age thats above 100

```python
df = df[(df['age'] >= 0) & (df['age'] <= 100)]
df.age
```

```python
print(df.shape)
df.head()
df.patientid
```

## Dropping PatientID and AppointmentID because they don't have benefits for me in next steps

```python
df = df.drop(columns=['patientid', 'appointmentid'])
df.head()
```

## Making sure the no_show columns data looks alike with "Yes" and "No" 

```python
df.no_show.value_counts()
df['no_show'] = df['no_show'].str.strip().str.title()
```

## Filtering handcap data so it become 0 or 1 

```python
df['handcap'] = df['handcap'].clip(upper=1)
df.handcap.value_counts()
```

## Dropping neighbourhood because it contain only 2 rows which isnt necessary for our analysis then making the new file

```python
df.drop(df[df['neighbourhood'] == 'ILHAS OCEÂNICAS DE TRINDADE'].index, inplace=True)
df.to_csv("preprocessed_data.csv", index=False)
```
